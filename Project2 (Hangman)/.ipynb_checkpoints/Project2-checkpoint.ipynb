## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to win the game.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible. In this way,all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, we have given length of word + 2 chances. For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.

```python
words='''apple banana mango strawberry
orange grape pineapple apricot lemon coconut watermelon 
cherry papaya berry peach lychee muskmelon'''
```

```python
import random
```

```python
word=words.split()
word
```

```python
while True:
    
    word_chosen=random.choice(word)
    no_chances=len(word_chosen)+2
    letter_guessed=''
    
    print(f'You have number of tries: {no_chances}')
    print('_ '*len(word_chosen))


    for i in range (no_chances):
        char_guessed=input('\nEnter the character you want: ')
                
        if not char_guessed.isalpha():
            print('Please enter a character not a number')
            continue

        elif char_guessed in letter_guessed:
            print('Already Guessed')
            continue

        elif len(char_guessed)>1:
            print('Please enter a single character ')
            continue


        if char_guessed in word_chosen:
            k=word_chosen.count(char_guessed)
            for i in range(k):
                letter_guessed+=char_guessed

        if len(word_chosen)==len(letter_guessed):
            print('Great Work You Win')
            print(f'The word chosen is : {word_chosen}')
            break

        else:
            for c in word_chosen:
                if c in letter_guessed:
                    print(c,end=' ')
                    
                else:
                    print('_ ',end=' ')


    else:
        print(f'Game Over The Correct Word Chosen Is {word_chosen}')

    if input('Do You Want To Play Again ?? (y/n): ')=='n':
        break

    
        
```
