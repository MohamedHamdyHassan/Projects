### Description 
> #### In this game, the user gets the first chance to pick the option among Rock, paper and scissors. After that computer select from the remaining two choices(randomly), the winner is decided as per the rules.
- Rock vs paper-> paper wins
- Rock vs scissor-> Rock wins
- paper vs scissor-> scissor wins.

```python
import random
```

```python
choices=['Rock','Paper','Scissor']

while True:
    
    user_choice=int(input('enter ur choice: '))

    if user_choice>3 or user_choice<1:
        print('please enter a valid number (1,2,3): ')
        continue

    user_choice_name=choices[user_choice-1]

    comp_choice=random.randint(1,3)
    if comp_choice==user_choice:
        comp_choice=random.randint(1,3)

    comp_choice_name=choices[comp_choice-1]

    print(f'The user choice is : {user_choice_name}')
    print(f'The computer choice is : {comp_choice_name}')

    print(user_choice_name,'V/S',comp_choice_name)

    if (user_choice==1 and comp_choice==2) or ((user_choice==2 and comp_choice==1)):
        result='Paper'

    elif (user_choice==1 and comp_choice==3) or ((user_choice==3 and comp_choice==1)):
        result='Rock'

    else:
        result='Scissor'

    if user_choice_name==result:
        print('Great Work You Win')

    else:
        print('You lost')
    
    print(f'The Answer is: {result}')
    
    if input('Do you want to play again ?? (y/n):')=='n':
        break
```
